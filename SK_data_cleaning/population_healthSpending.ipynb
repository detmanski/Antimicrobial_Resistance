```python
## Imports
import pandas as pd
import numpy as np
```

```python
# Establishes the base filepath to find and write the data
## NOTE: This will be different for each computer
filepath = "C:/Users/kronh/OneDrive/Documents/UofTCoding_bootcamp/project_3_data"
```

Source:
https://apps.who.int/nha/database

```python
# Reads the data into a DataFrame
df = pd.read_csv(f"{filepath}/data_csv/GHED_data.csv")

# Removes unnecessary columns
df = df[['country', 'code', 'income', 'year', 'che_usd', 'pop', 'che_pc_usd']]

# Formats the health spending per capita and population columns
df.rename(columns={'che_usd': 'health_spending_mil_USD','che_pc_usd':'health_spending_per_capita_USD', 'pop':'population_thousands'}, inplace=True)
df.replace(",","", regex=True, inplace=True)
df[['health_spending_mil_USD', 'health_spending_per_capita_USD', 'population_thousands']] = df[['health_spending_mil_USD', 'health_spending_per_capita_USD', 'population_thousands']].astype(float)

# Creates an ID column
df.insert(0, 'id', range(1,1+len(df)))

# Imports the countries csv in order to add region_id and country_id columns to the countries DataFrame
countries_df = pd.read_csv(f"{filepath}/data_csv/countries.csv")

# joins the DataFrames to add region_id and country_id to the spending/ population DataFrame
joined_df = pd.merge(df, countries_df, how='left', on='country')

# Renames the id columns
joined_df.rename(columns={'id_x':'id', 'id_y':'country_id'}, inplace=True)

# Moves the FK columns to the front of the DataFrame
joined_df = joined_df[['id', 'country_id', 'region_id', 'country', 'code', 'income', 'year',
       'health_spending_mil_USD', 'population_thousands',
       'health_spending_per_capita_USD']]
```

```python
# Exports the DataFrame to a csv for input into a database
joined_df.to_csv(f"{filepath}/data_csv/spending_pop.csv", index=False)
```
